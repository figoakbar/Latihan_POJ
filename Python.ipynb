```python
#import Library Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import Library Numpy
import numpy as np
#Menghitung jumlah data yg berjarak dari 0 sampai 2phi
x = np.linspace(0,2*np.pi, 100)
#kemudian menghitung cos si x
cos_x = np.cos(x)
```

```python
#membuat line port
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x)
```

```python
#membuat sumbu x dan y  memmiliki rasio yg sama
fig, ax = plt.subplots()
_= ax.plot(x, cos_x)
_= ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x, markersize=20, linestyle='-.',
            color='yellow', label='cos')
_ = ax.set_aspect('equal')
_ = ax.legend()
```
